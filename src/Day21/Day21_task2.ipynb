Types of Machine Learning

1. Supervised Learning
   Supervised Learning is a type of machine learning where the data already has labels or correct answers.
   The model learns from input-output examples and tries to predict the correct output for new data.

2. Unsupervised Learning
   Unsupervised Learning is used when the data has no labels or answers.
   The model tries to find hidden patterns, similarities, or clusters in the data by itself.  

3. Reinforcement Learning
   Reinforcement Learning is where a system learns by trial and error.
   It receives rewards for correct actions and penalties for wrong actions, and improves over time.


## Comparison of Machine Learning Types

| Learning Type | Data Has Labels? | Main Goal | Real-World Examples |
|---------------|------------------|-----------|---------------------|
| Supervised Learning | Yes | Predict outputs or classify data using labeled examples | Spam Detection, House Price Prediction |
| Unsupervised Learning | No | Discover hidden patterns or group similar data | Customer Segmentation, Grouping News Articles |
| Reinforcement Learning | No (uses rewards instead of labels) | Learn the best actions through rewards and penalties | Self-Driving Cars, Game Playing AI |
